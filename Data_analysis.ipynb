```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('dataset/dataset.csv')
df.head()
```

### Year-wise analysis of each month

```python
fig, axs = plt.subplots(nrows=5, ncols=4, figsize=(15, 10))

for i, ax in zip(range(2001, 2021), axs.flatten()):
    x = []
    y = []

    df_year = df[df['year'] == i]

    for j in range(1, 13):
        df_month = df_year[df_year['month'] == j]
        x.append(j)
        y.append(df_month['rainfall'].mean())

    ax.plot(x, y, linestyle='--', marker='o', color='b')
    ax.set_xlabel(str(i)+' - months')
    ax.set_ylabel('Rainfall (mm)')
    ax.grid()

plt.tight_layout()
plt.show()
```

## Analysis of all year

#### Maximum

```python
x = []
y = []

for i in range(2001,2021):

    df_year = df[df['year'] == i]
    x.append(i)
    y.append(df_year['rainfall'].mean())

plt.figure(figsize=(15,6))
plt.plot(x,y, linestyle='-', marker='o', color='b')
plt.xlabel("Years")
plt.ylabel('Rainfall (mm)')
plt.xticks(np.arange(min(x), max(x)+1, 1))
plt.grid()
plt.show()
```

#### Average

```python
x = []
y = []

for i in range(2001,2021):

    df_year = df[df['year'] == i]
    x.append(i)
    y.append(df_year['rainfall'].max())

plt.figure(figsize=(15,6))
plt.plot(x,y, linestyle='-', marker='o', color='b')
plt.xlabel("Years")
plt.ylabel('Rainfall (mm)')
plt.xticks(np.arange(min(x), max(x)+1, 1))
plt.grid()
plt.show()
```

#### Minimum

```python
x = []
y = []

for i in range(2001,2021):

    df_year = df[df['year'] == i]
    x.append(i)
    y.append(df_year['rainfall'].min())

plt.figure(figsize=(15,6))
plt.plot(x,y, linestyle='-', marker='o', color='b')
plt.xlabel("Years")
plt.ylabel('Rainfall (mm)')
plt.xticks(np.arange(min(x), max(x)+1, 1))
plt.grid()
plt.show()
```
